```python
def total_distance(l1: list, l2: list) -> int:
    res = 0
    while l1 and l2:
        min_l1 = min(l1)
        min_l2 = min(l2)
        
        l1.remove(min_l1)
        l2.remove(min_l2)
        
        res += abs(min_l1-min_l2)
        
    return res

total_distance(l1,l2)
                
```

```python
def similarity_score(l1: list, l2: list) -> int:
    similarity_dict = {}
    l1 = set(l1)
    for number in l1:
        similarity_dict[number] = l2.count(number)
    return sum({key*value for key, value in similarity_dict.items()})

similarity_score(l1,l2)
```
